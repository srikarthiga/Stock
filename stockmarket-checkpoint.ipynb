```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from statsmodels.tsa.stattools import adfuller
from pmdarima import auto_arima
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Load dataset
file_path = "stock_data.csv"
data = pd.read_csv(file_path, parse_dates=['Date'], index_col='Date')

# Set frequency to business days to ensure continuous time index
data = data.asfreq('B')  # B = Business day

# Forward-fill missing values created by asfreq()
data['Close'] = data['Close'].ffill()

# ADF test for stationarity
result = adfuller(data['Close'])
adf_stat = result[0]
p_value = result[1]

print(f"ADF Statistic: {adf_stat}")
print(f"p-value: {p_value}")

# If not stationary, difference the data
if p_value > 0.05:
    data_diff = data['Close'].diff().dropna()
else:
    data_diff = data['Close']

# Auto ARIMA to select best (p,d,q)
model_auto = auto_arima(data_diff, seasonal=False, trace=False, error_action='ignore', suppress_warnings=True)
order = model_auto.order
print(f"Best ARIMA order: {order}")

# Fit ARIMA model
model = ARIMA(data['Close'], order=order)
model_fit = model.fit()

# Forecast: last 30 actual + next 10 predicted
forecast = model_fit.predict(start=len(data)-30, end=len(data)+10)
forecast_index = pd.date_range(start=data.index[-30], periods=41, freq='B')

# Evaluate on last 30 points
actual = data['Close'][-30:]
predicted = forecast[:30]

mae = mean_absolute_error(actual, predicted)
rmse = np.sqrt(mean_squared_error(actual, predicted))
print(f"\nEvaluation on last 30 days:\nMAE: {mae:.2f}, RMSE: {rmse:.2f}")

# Plot actual and forecast
plt.figure(figsize=(12, 6))
plt.plot(data['Close'], label='Actual', color='blue')
plt.plot(forecast_index, forecast, label='Forecast', color='red')
plt.axvline(x=data.index[-1], color='gray', linestyle='--')  # mark forecast start
plt.title("ARIMA Forecast")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load and preprocess data
data = pd.read_csv("stock_data.csv", parse_dates=['Date'], index_col='Date')
data = data.asfreq('D')  # Ensure frequency is set
data['Close'] = data['Close'].ffill()  # Forward-fill missing values

# Define SARIMA model
sarima_model = SARIMAX(data['Close'],
                       order=(1, 1, 1),
                       seasonal_order=(1, 1, 1, 12),
                       enforce_stationarity=False,
                       enforce_invertibility=False)

# Fit the model
sarima_result = sarima_model.fit()

# Summary
print(sarima_result.summary())

# Forecasting the next 30 time steps
n_steps = 30
sarima_forecast = sarima_result.get_forecast(steps=n_steps)

# Get confidence intervals
conf_int = sarima_forecast.conf_int()
forecast_index = pd.date_range(start=data.index[-1] + pd.Timedelta(days=1), periods=n_steps, freq='D')

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Close'], label='Observed')
plt.plot(forecast_index, sarima_forecast.predicted_mean, label='Forecast', color='red')
plt.fill_between(forecast_index,
                 conf_int.iloc[:, 0],
                 conf_int.iloc[:, 1],
                 color='pink', alpha=0.3)
plt.title("SARIMA Forecast")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
data = pd.read_csv("stock_data.csv") 
data = data[['Date', 'Close']]
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data[['Close']])
train_size = int(len(scaled_data) * 0.8)
train_data = scaled_data[:train_size]

x_train = []
y_train = []

for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])

x_train = np.array(x_train)
y_train = np.array(y_train)

x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(LSTM(units=50))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(x_train, y_train, epochs=5, batch_size=32)
test_data = scaled_data[train_size - 60:]

x_test = []
y_test = data['Close'].values[train_size:]

for i in range(60, len(test_data)):
    x_test.append(test_data[i-60:i, 0])

x_test = np.array(x_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)
plt.figure(figsize=(10, 5))
plt.plot(y_test, color='blue', label='Real Stock Price')
plt.plot(predictions, color='red', label='Predicted Stock Price')
plt.title('Stock Price Prediction using LSTM')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM

# 1. Load the dataset
df = pd.read_csv("stock_data.csv")
df = df[['Date', 'Close']]
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df.dropna(inplace=True)

# 2. Scale the data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df[['Close']])

# 3. Prepare training data
sequence_length = 60
X = []
y = []

for i in range(sequence_length, len(scaled_data)):
    X.append(scaled_data[i-sequence_length:i, 0])
    y.append(scaled_data[i, 0])

X = np.array(X)
y = np.array(y)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))  # (samples, time steps, features)

# 4. Build LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X.shape[1], 1)))
model.add(LSTM(units=50))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')

# 5. Train the model
model.fit(X, y, epochs=5, batch_size=32, verbose=1)

# 6. Make predictions
predicted_prices = model.predict(X)
predicted_prices = scaler.inverse_transform(predicted_prices)
actual_prices = scaler.inverse_transform(y.reshape(-1, 1))

# 7. Plot the results
plt.figure(figsize=(14, 5))
plt.plot(actual_prices, color='blue', label='Actual Stock Price')
plt.plot(predicted_prices, color='red', label='Predicted Stock Price by LSTM')
plt.title('Stock Price Prediction using LSTM')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.grid()
plt.show()
```
